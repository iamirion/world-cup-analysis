```python
def find_winner(row):
    if row['Home Goals'] > row['Away Goals']:
        return row['Home Team']
    elif row['Away Goals'] > row['Home Goals']:
        return row['Away Team']
    else:
        return 'Draw'
```

```python
df['winner']= df.apply(find_winner, axis=1)
df
```
